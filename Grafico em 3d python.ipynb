```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Criar figura e eixo 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Criar dados para plotagem da espiral
dim_z = np.linspace(0, 18, 1000)
dim_x = np.cos(dim_z)
dim_y = np.sin(dim_z)

# Plotar espiral de linha no plano 3D
ax.plot3D(dim_x, dim_y, dim_z, 'purple')

# Configurar título e rótulos dos eixos
ax.set_title('Espiral de linha no Plano 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

# Modificar ângulo de visualização do gráfico
ax.view_init(10, 45)

# Mostrar gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

dim_z = np.linspace(0, 15, 1000)
dim_x = np.cos(dim_z)
dim_y = np.sin(dim_z)

ax.plot3D(dim_x, dim_y, dim_z, 'purple')

pontos_z = 15 * np.random.random(120)
pontos_x = np.cos(pontos_z) + 0.1 * np.random.randn(120)
pontos_y = np.sin(pontos_z) + 0.1 * np.random.randn(120)

ax.scatter3D(pontos_x, pontos_y, pontos_z, c=pontos_z, cmap='Spectral')

ax.set_title('Espiral de linha e pontos no Plano 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

# [view_init] Modifica o ângulo de visualização do gráfico
ax.view_init(50, 35)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='coolwarm')

ax.set_title('Gráficos de contorno no Espaço 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

ax.view_init(70, 35)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Criando a figura e o eixo 3D
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection="3d")

# Definindo o título e rótulos dos eixos
ax.set_title('Ambiente de Plotagem 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

# Exibindo o gráfico
plt.show()

# Plotando uma espiral 3D
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

# Gerando os dados
dim_z = np.linspace(0, 18, 1000)
dim_x = np.cos(dim_z)
dim_y = np.sin(dim_z)

# Plotando a linha
ax.plot3D(dim_x, dim_y, dim_z, 'purple')

# Definindo o título e rótulos dos eixos
ax.set_title('Espiral de linha no Plano 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

# Exibindo o gráfico
plt.show()

# Plotando uma espiral 3D com pontos
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

# Gerando os dados da linha
dim_z = np.linspace(0, 15, 1000)
dim_x = np.cos(dim_z)
dim_y = np.sin(dim_z)

# Gerando os dados dos pontos
pontos_z = 15 * np.random.random(120)
pontos_x = np.cos(pontos_z) + 0.1 * np.random.randn(120)
pontos_y = np.sin(pontos_z) + 0.1 * np.random.randn(120)

# Plotando a linha e os pontos
ax.plot3D(dim_x, dim_y, dim_z, 'purple')
ax.scatter3D(pontos_x, pontos_y, pontos_z, c=pontos_z, cmap='Spectral')

# Definindo o título e rótulos dos eixos
ax.set_title('Espiral de linha e pontos no Plano 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

# Exibindo o gráfico
plt.show()
```
